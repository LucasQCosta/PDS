```python
#Rodando AÃºdio WAV
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from IPython.display import Audio
import torch

signal, samplerate = sf.read('audio.wav')
'''
signal => sinal armazenado em uma numpy
samplerate => a taxa de amostragem do sinal
'''

time = torch.arange(0, len(signal)) / samplerate

print(f'FrequÃªncia de Amostragem: {samplerate} Hz')
plt.title('Forma da Onda')
plt.plot(time, signal)
plt.xlabel("Tempo (s)");
plt.ylabel("Amplitude (db)");
plt.grid()
plt.show()


Audio("audio.wav")
```
